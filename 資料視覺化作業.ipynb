> 載入函式庫

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

import plotly.offline
import plotly.express as px
import plotly.graph_objects as go
import plotly.subplots
```

> 讀入 保險資料 檔案（insurance.csv）

```python
df = pd.read_csv('insurance.csv')
```

### **練習 1**：繪製「年齡」直方圖

```python
sns.distplot(df["Age"], kde=False)
```

### **練習2**：利用「車齡」做分類，繪製三種車齡下之「合約時長」重疊直方圖

```python
two_years = df[df["Vehicle_Age"] == "> 2 Years"]
one_to_two_year = df[df["Vehicle_Age"] == "1-2 Year"]
one_year = df[df["Vehicle_Age"] == "< 1 Year"]

sns.distplot(one_to_two_year["Vintage"], kde=False, color="g")
sns.distplot(one_year["Vintage"], kde=False, color="b")
sns.distplot(two_years["Vintage"], kde=False, color="r")
```

### **練習3**：請繪製「有無興趣」人數分佈

```python
sns.countplot(x="Response", data=df)
```

### **練習4**：請繪製「不同車齡下」，有無興趣之人數分佈

```python
sns.countplot(x="Response", hue="Vehicle_Age", data=df)
```

### **練習5**

> 5.1 「合約時長」盒鬚圖

```python
sns.boxplot(y="Vintage", data=df)
```

> 5.2 「不同車齡下」，「年齡」的盒鬚圖

```python
sns.boxplot(x="Vehicle_Age", y="Age", data=df)
```

### **練習6**

> 6.1 「不同車齡下」，「年齡」的盒鬚圖，並設定「車齡」的呈現順序

```python
sns.boxplot(x="Vehicle_Age", y="Age", order=["< 1 Year", "1-2 Year", "> 2 Years"], data=df)
```

> 6.2 「不同車齡下」，分別呈現有興趣跟無興趣者的年保費盒鬚圖

```python
sns.boxplot(x="Vehicle_Age", y="Annual_Premium", hue="Response", data=df)
```
